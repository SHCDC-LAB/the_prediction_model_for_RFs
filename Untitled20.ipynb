```python
# 导入数据
import csv
import os
import multiprocessing
import pandas as pd
import numpy as np
from rdkit.Chem import PandasTools
from supervised.automl import AutoML
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
from math import sqrt

data = pd.read_csv('D:\PPpython\RIpredition\dataRIs.csv', names=['Smiles','Name','RF']) 
PandasTools.AddMoleculeColumnToFrame(data,smilesCol='Smiles')
from padelpy import padeldescriptor
PandasTools.AddMoleculeColumnToFrame(data,smilesCol='Smiles')
PandasTools.WriteSDF(data,'output.sdf',idName=None,molColName='ROMol',properties=list(data.columns),allNumeric=False)
padeldescriptor(mol_dir='output.sdf',d_file='descriptors.csv',d_2d=True,d_3d=False,retainorder=True,threads=1)
padeldescriptor(mol_dir='output.sdf',d_file='fingerprints.csv',fingerprints=True,retainorder=True,threads=1)
# 训练集第一列替换为RT
df1 = pd.read_csv('descriptors.csv')
df1['Name'] = data['RF']
df1.rename(columns={'Name':'RF'}, inplace=True)
df1.to_csv('descriptors.csv', index=False)
df2 = pd.read_csv('fingerprints.csv')
df2['Name'] = data['RF']
df2.rename(columns={'Name':'RF'}, inplace=True)
df2.to_csv('fingerprints.csv', index=False)
# 读取数据
df = pd.read_csv("descriptors.csv")

# 替换无效值 '#DIV/0!' 为 NaN，并删除包含 NaN 的行
df.replace('#DIV/0!', np.nan, inplace=True)
df.dropna(inplace=True)  # 删除包含 NaN 的行

# 将所有列转换为数值类型，错误的转换会变成 NaN
df = df.apply(pd.to_numeric, errors='coerce')

# 再次清理数据，删除因转换失败导致的 NaN
df.dropna(inplace=True)

# 分离特征和目标变量
X = df.loc[:, df.columns != 'RF']
y = df["RF"]

# 确保目标变量 y 为数值类型
y = pd.to_numeric(y, errors='coerce')

# 分割数据集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# 定义AutoML对象，限制CPU负载
automl_desc = AutoML(
    algorithms=["Xgboost"],         # 指定算法
    mode="Perform",                 # 执行模式
    results_path="RTprediction_pos_2",# 结果保存路径
    eval_metric="rmse",             # 评估指标
    total_time_limit=3600,          # 总时间限制
    n_jobs=1                        # 设置并行线程数，减少CPU负载
)

# 训练模型
automl_desc.fit(X_train, y_train)

# 结果评价
pred_train = automl_desc.predict(X_train)
pred_test = automl_desc.predict(X_test)
rmse_train = sqrt(mean_squared_error(y_train, pred_train))
rmse_test = sqrt(mean_squared_error(y_test, pred_test))

# 输出RMSE
print("Train RMSE:", rmse_train)
print("Test RMSE:", rmse_test)

data = pd.read_csv('D:\\PPpython\\RIpredition\\nontargetPFAS.csv', names=['Smiles', 'Name'], encoding='ISO-8859-1')
PandasTools.AddMoleculeColumnToFrame(data, smilesCol='Smiles')
from padelpy import padeldescriptor
PandasTools.AddMoleculeColumnToFrame(data,smilesCol='Smiles')
PandasTools.WriteSDF(data,'output.sdf',idName=None,molColName='ROMol',properties=list(data.columns),allNumeric=False)
padeldescriptor(mol_dir='output.sdf',d_file='descriptors.csv',d_2d=True,d_3d=False,retainorder=True,threads=1)
df = pd.read_csv("D:\\PPpython\\RIpredition\\nontargetPFAS.csv")
automl = AutoML(results_path="AutoML_Desc_01")
## Split training data and target Name
df_train = df.loc[:, df.columns!='Name']
# Predict unknow features
pred_test = automl.predict(df_train)
pred_test = pd.DataFrame(pred_test)
pred_test.to_csv('D:\\PPpython\\RIpredition\\train2_pred.csv',index=False)
```
